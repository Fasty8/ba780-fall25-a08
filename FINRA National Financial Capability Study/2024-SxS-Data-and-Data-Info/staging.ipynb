```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset (replace with your actual data)
df = pd.DataFrame({
    'age': [15, 22, 25, 33, 40, 45, 52, 60, 67, 70, 75, 80, 85, 90, 22, 25, 25, 33, 33, 52]
})
```

```python
# Count occurrences of each age
age_counts = df['age'].value_counts().sort_index()

# Plot
sns.set_style("whitegrid")
sns.barplot(x=age_counts.index, y=age_counts.values, palette='deep')

# Labels and title
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# -----------------
# Create dummy dataset
# -----------------
np.random.seed(42)
age_groups = np.random.choice(
    ["18-25", "26-35", "36-45", "46-60", "60+"], 
    size=200
)
knowledge_levels = np.random.randint(1, 8, size=200)  # scale 17

df = pd.DataFrame({
    "Age Group": age_groups,
    "Financial Knowledge": knowledge_levels
})

# -----------------
# 1. Boxplot (distribution per age group)
# -----------------
plt.figure(figsize=(8,5))
sns.boxplot(x="Age Group", y="Financial Knowledge", data=df, palette="pastel")
plt.title("Financial Knowledge by Age Group (Boxplot)")
plt.show()

# -----------------
# 2. Violin Plot (distribution shape)
# -----------------
plt.figure(figsize=(8,5))
sns.violinplot(x="Age Group", y="Financial Knowledge", data=df, palette="muted")
plt.title("Financial Knowledge by Age Group (Violin Plot)")
plt.show()

# -----------------
# 3. Bar Plot (mean values with error bars)
# -----------------
plt.figure(figsize=(8,5))
sns.barplot(x="Age Group", y="Financial Knowledge", data=df, palette="deep", ci="sd")
plt.title("Average Financial Knowledge by Age Group (Bar Plot)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# -----------------
# Create dummy dataset
# -----------------
np.random.seed(42)
age_groups = np.random.choice(
    ["18-25", "26-35", "36-45", "46-60", "60+"], 
    size=200
)

# Convert numeric scale (17) into categorical levels
knowledge_levels = np.random.randint(1, 8, size=200)
knowledge_cats = pd.cut(
    knowledge_levels,
    bins=[0, 3, 5, 7],
    labels=["Low", "Medium", "High"]
)

df = pd.DataFrame({
    "Age Group": age_groups,
    "Financial Knowledge": knowledge_cats
})

# -----------------
# Create cross-tab (counts per category)
# -----------------
ct = pd.crosstab(df["Age Group"], df["Financial Knowledge"])

# -----------------
# Plot heatmap
# -----------------
plt.figure(figsize=(8,6))
sns.heatmap(ct, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Age Group vs. Financial Knowledge")
plt.ylabel("Age Group")
plt.xlabel("Financial Knowledge Level")
plt.show()
```
