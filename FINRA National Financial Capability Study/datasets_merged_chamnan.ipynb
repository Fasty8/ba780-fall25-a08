## FinLit Calculation

```python
# ======================================
# BUCKET 1: Critical Stability (max 5 each)
# ======================================
emergency_funds_w = {  # J5 — Emergency or rainy day funds
    1: 5,  # Yes
    2: 1,  # No
    98: 0, 99: 0, "No Response Given": 0
}

retirement_planning_w = {  # J8 — Tried to figure out retirement savings
    1: 5,
    2: 1,
    98: 0, 99: 0, "No Response Given": 0
}

health_insurance_w = {  # H1 — Covered by health insurance
    1: 5,
    2: 1,
    98: 0, 99: 0, "No Response Given": 0
}

# ======================================
# BUCKET 2: Core Access & Assets (max 3 each)
# ======================================
checking_account_w = {  # B1
    1: 3,
    2: 1,
    98: 0, 99: 0, "No Response Given": 0
}

savings_account_w = {  # B2
    1: 3,
    2: 1,
    98: 0, 99: 0, "No Response Given": 0
}

other_investments_w = {  # B14 — Non-retirement investments
    1: 3,
    2: 1,
    98: 0, 99: 0, "No Response Given": 0
}

retirement_planning_w = {  # C5 — Contribute to retirement account
    1: 3,
    2: 1,
    98: 0, 99: 0, "No Response Given": 0
}

# ======================================
# BUCKET 3: Credit Health & Debt Behavior (max 3 each)
# ======================================
overdraw_w = {  # B4 — Overdraw checking?
    2: 3,  # No (best)
    1: 1,  # Yes
    98: 0, 99: 0, "No Response Given": 0
}

credit_card_balance_interest_w = {  # F2_2 — Carried balance & charged interest
    2: 3,  # No (best)
    1: 1,  # Yes
    98: 0, 99: 0, "No Response Given": 0
}

num_credit_cards_w = {  # F1 — Number of credit cards
    2: 3,  # 2–3 (ideal)
    3: 3,  # 4–8
    1: 2,  # 1
    4: 1,  # 9–12
    5: 0,  # 13–20
    6: 0,  # >20
    7: 1,  # None
    98: 0, 99: 0, "No Response Given": 0
}


# ======================================
# COMBINED WEIGHTS DICTIONARY
# ======================================
WEIGHTS = {
    # Bucket 1: Critical Stability
    "Emergency Funds": emergency_funds_w,
    "Retirement Planning": retirement_planning_w,
    "Health Insurance": health_insurance_w,

    # Bucket 2: Core Access & Assets
    "Checking Account": checking_account_w,
    "Savings Account": savings_account_w,
    "Other Investments": other_investments_w,
    "Other Retirement Accounts": retirement_planning_w,

    # Bucket 3: Credit Health & Debt Behavior
    "Overdraw": overdraw_w,
    "Credit Card Balance Interest": credit_card_balance_interest_w,
    "Number of Credit Cards": num_credit_cards_w,
}

# GRAND TOTAL MAX = 15 + 12 + 9 + 10 = 46

for col, weight_dict in WEIGHTS.items():
    # financial_literacy[f"{col}_score"] = financial_literacy[col].map(weight_dict)
    cleaned_df2024[f"{col}_score"] = cleaned_df2024[col].map(weight_dict)

```

```python
cleaned_df2024 = cleaned_df2024.replace(column_values_mapping)
```

```python
cleaned_df2024["Total_Score"] = cleaned_df2024[[f"{col}_score" for col in WEIGHTS.keys()]].sum(axis=1)

conditions = [
    (cleaned_df2024["Total_Score"] >= 40),                                      # Very High
    (cleaned_df2024["Total_Score"] >= 34) & (cleaned_df2024["Total_Score"] < 40),  # High
    (cleaned_df2024["Total_Score"] >= 27) & (cleaned_df2024["Total_Score"] < 34),  # Moderately High
    (cleaned_df2024["Total_Score"] >= 20) & (cleaned_df2024["Total_Score"] < 27),  # Average
    (cleaned_df2024["Total_Score"] >= 13) & (cleaned_df2024["Total_Score"] < 20),  # Moderately Low
    (cleaned_df2024["Total_Score"] >= 6)  & (cleaned_df2024["Total_Score"] < 13),  # Low
    (cleaned_df2024["Total_Score"] >= 0)  & (cleaned_df2024["Total_Score"] < 6),    # Very Low
    (cleaned_df2024["Total_Score"] == 0)                             # No Response Given
]

choices = [
    "Very High", 
    "High", 
    "Moderately High", 
    "Average", 
    "Moderately Low", 
    "Low", 
    "Very Low", 
    "None"
]

cleaned_df2024["FinLit_Level"] = np.select(conditions, choices, default="No Response Given")

```
